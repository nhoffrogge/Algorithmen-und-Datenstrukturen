# Big O Notation
In dieser Lektion werden wir besprechen, wie die Syntax der Big-O Notation funktioniert und wie wir Algorithmen mit der Big-O Notation beschreiben können!

Wir haben zuvor die folgenden Funktionen besprochen:

```python
# Erste Funktion (Beachte die Verwendung von range)
def sum1(n):
    '''
    Nimm eine Eingabe von n und gib die Summe der Zahlen von 0 bis n zurück.
    '''
    final_sum = 0
    for x in range(n+1): 
        final_sum += x
        
    # O (n)
    
    return final_sum
```

```python
def sum2(n):
    """
    Nimm eine Eingabe von n und gib die Summe der Zahlen von 0 bis n zurück.
    """
    return (n*(n+1))/2
```

Nun wollen wir eine Notation entwickeln, um die Effizienz dieser beiden Algorithmen objektiv zu vergleichen. Ein guter Ausgangspunkt wäre es, die Anzahl der Zuweisungen zu vergleichen, die jeder Algorithmus vornimmt.

Die ursprüngliche **sum1** Funktion erstellt eine Zuordnung **n+1** mal, wir können dies an der bereichsbasierten Funktion erkennen. Das bedeutet, dass es die final_sum Variable n+1 mal zuweist. Wir können dann sagen, dass für ein Problem der Größe n (in diesem Fall nur eine Zahl n) diese Funktion 1+n Schritte macht.

Diese **n-** Notation ermöglicht es uns, Lösungen und Algorithmen relativ zur Größe des Problems zu vergleichen, da sum1(10) und sum1(100000) sehr unterschiedliche Laufzeiten benötigen würden, aber den gleichen Algorithmus verwenden. Wir können auch feststellen, dass mit zunehmender Größe von n die **+1** nicht viel Wirkung hat. Beginnen wir also mit der Diskussion, wie man eine Syntax für diese Notation erstellt.
________

Nun werden wir darüber diskutieren, wie wir diese Notation und Idee formalisieren können.

Die Big-O-Notation beschreibt, *wie schnell die Laufzeit relativ zur Eingabe wächst, wenn die Eingabe beliebig groß wird*.

Betrachten wir einige dieser Punkte genauer:

* Bitte denk daran, wir möchten vergleichen, wie schnell die Laufzeit wächst, und nicht die genauen Laufzeiten, da diese je nach Hardware variieren können.

* Da wir eine Vielzahl von Eingabegrößen vergleichen wollen, geht es uns nur um die Laufzeitentwicklung *relativ* zur Eingabe. Deshalb verwenden wir **n** für die Notation.

* Da n beliebig groß wird, kümmern wir uns nur um Terme, die am schnellsten wachsen, wenn n groß wird, deswegen ist die Big-O-Analyse auch als **asymptotische Analyse** bekannt.


Als Syntax kann man sagen, dass sum1() **O(n)** ist, da seine Laufzeit linear mit der Eingabegröße wächst. In der nächsten Lektion werden wir auf spezifischere Beispiele für verschiedene O()-Typen und Beispiele eingehen. Zum Abschluss dieser Lektion werden wir das Potenzial für große Laufzeitunterschiede von Big-O-Funktionen aufzeigen.

Damit wir die Laufzeiten vergleichen können, plotten wir die Laufzeit im Vergleich zum Big-O. Wir verwenden einfach [matplotlib](http://matplotlib.org/) für die nachfolgende Darstellung. (Kümmere dich bitte nicht um die Verwendung von matplotlib, das ist für diesen Teil irrelevant.)

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

# Laufzeitvergleiche einrichten
n = np.linspace(1,10,1000)
labels = ['Constant','Logarithmic','Linear','Log Linear','Quadratic','Cubic','Exponential']
big_o = [np.ones(n.shape),np.log(n),n,n*np.log(n),n**2,n**3,2**n]

# Plot-Einstellung
plt.figure(figsize=(12,10))
plt.ylim(0,50)

for i in range(len(big_o)):
    plt.plot(n,big_o[i],label = labels[i])


plt.legend(loc=0)
plt.ylabel('Relative Laufzeit')
plt.xlabel('n')
```

Bitte beachte, was für einen Unterschied der Big-O Wirkungsgrad für den gleichen n-Wert im Sinne der prognostizierten Laufzeit macht! Natürlich sind wir bestrebt, Algorithmen zu wählen, die sich von jedem exponentiellen, quadratischen oder kubischen Verhalten fernhalten!
