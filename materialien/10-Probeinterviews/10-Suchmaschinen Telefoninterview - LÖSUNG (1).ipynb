## Technische Fragen

**Beantworte die folgende Frage in der untenstehenden *Markdown* Zelle. Es ist wichtig zu beachten, dass die Zelle unten KEINE Syntaxhervorhebung hat, es ist üblich, in einem Telefoninterview einen Texteditor zu erhalten, der nichts anderes als einfache Textunterstützung hat**.

1. Schreibe eine Funktion, die die N-te Fibonacci-Zahl berechnet.

## Lösung

Es gibt viele Möglichkeiten, diese Frage zu beantworten, du könntest aufgefordert werden, sie auf verschiedene Weise zu lösen und einige Vor- und Nachteile jeder Art zu diskutieren. Nachfolgend sind verschiedene Lösungen aufgeführt

```python
## Beispiel 1: Verwendung der Schleifen-Technik
def fib(n):
    
    a,b = 1,1
    for i in range(n-1):
        a,b = b,a+b
    return a

print fib(7)
    
# Verwendung von Rekursion       
def fibR(n):
    if n==1 or n==2:
        return 1
    return fib(n-1)+fib(n-2)

print fibR(7)
 
## Beispiel 3: Verwendung von Generatoren
a,b = 0,1
def fibI():
    global a,b
    while True:
        a,b = b, a+b
        yield a
f=fibI()
f.next()
f.next()
f.next()
f.next()
f.next()
f.next()
print f.next()

 
## Beispiel 4: Verwendung der Memoisierung
def memoize(fn, arg):
    memo = {}
    if arg not in memo:
        memo[arg] = fn(arg)
    return memo[arg]
 
## fib() wie in Beispiel 1 geschrieben.
fibm = memoize(fib,7)
print fibm
 
## Beispiel 5: Memoisierung als Dekorateur verwenden
class Memoize:
    def __init__(self, fn):
        self.fn = fn
        self.memo = {}
    def __call__(self, arg):
        if arg not in self.memo:
            self.memo[arg] = self.fn(arg)
            return self.memo[arg]
 
@Memoize
def fib(n):
    a,b = 1,1
    for i in range(n-1):
        a,b = b,a+b
    return a
print fib(7)
```
